```python
#EDA(Exploratory Data Analysis)
```

```python
#GENDER
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_genderbasis=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_genderbasis)
```

```python
#From above graphs we can see that most of the buyers are females and even the purchasing power of females greater than man
```

```python
#AGE

ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age Group

sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
#From above the graphs we acn see that most of the buyers are of age group between 26-35 yrs female
```

```python
#STATE
#Total number of orders from top 10 states

sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#Total amount /Sales from top 10 States

sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
#From above graphs we can see that most of the orders are from uttar pradesh,Maharastra and karnataka respectively 
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(6,5)})
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True).head(10)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_state)
```

```python
#From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
#Occupation
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(6,5)})

```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Occupation',y='Amount',data=sales_state)
```

```python
#From above the we can see can see that most of the buyers are working in IT,Aviation,and healthcare sector
```

```python
#product category
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(30,10)})
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_Category',y='Amount',data=sales_state)
```

```python
#From above the graphs we can see that most of the sold product are from Food,Footwear,and Electronics Category
```

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_ID',y='Orders',data=sales_state)
```

```python
#top 10 most sold product (Same thing as Above)

fig1,ax1=plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
